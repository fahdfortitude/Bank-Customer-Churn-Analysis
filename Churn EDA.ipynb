```python
benchmark="""current_balance
previous_month_end_balance
average_monthly_balance_prevQ
average_monthly_balance_prevQ2
current_month_credit
previous_month_credit
current_month_debit
previous_month_debit
current_month_balance
previous_month_balance""".split('\n')
```

```python
fig, ax=plt.subplots(5,2, figsize=(12,10))
for col, axis in zip(benchmark, ax.ravel()):
    sns.distplot(churn[col], ax=axis, axlabel=f'Distribution of {col}')
plt.tight_layout()
```

All the transactional history data sets are skewed to the right, hence, it is extremely difficult to examine the distributions. As a personal rule of thumb, next line of action is to visualize 95% of the data sets, this is suppose to exclude the outliers from the visualizations and visual examination of the distribution easier.

```python
fig, ax=plt.subplots(5,2, figsize=(12,10))
for col, axis in zip(benchmark, ax.ravel()):
    per_90=churn[col][churn[col]<=np.percentile(churn[col],95)]
    sns.distplot(per_90, ax=axis, axlabel=f'Distribution of {col}')
plt.tight_layout()
```

The Foregoing distribution shows that despite the fact that some customers transactional history are in millons, most of the bank's customers transactional history is less than 25000 for all the benchmarks, in fact most seem to have less than 5000. Also some customers have negative balances up to 5000.
Because of the conspicuous presence of outliers, the benchmark columns will need to logged when preprocessing them for machine learning. 
It is important to note that for metrics with negative number the logarithimic equation is <br><br></br></br>
A = Log( Y +  abs(min(Y))+ 1)<br></br>
where Y is a benchmark column values and abs(min(Y)) is the absolute value of the maximum negative number
<br><br></br></br>

###### Benchmarks and churn

```python
# benchmark=['current_balance', 'current_month_credit', 'current_month_debit']

fig, ax =plt.subplots(5,2, figsize=(16,18))


for col, axis in zip(benchmark, ax.ravel()):
    per_95=np.round(np.percentile(churn[col],95))
    sns.boxplot(x='churn', y=col, data=churn[churn[col]<=per_95], ax=axis)
    axis.set_xticklabels(labels=["Didn't Churn",'Churned'])
    axis.set(xlabel='')
    axis.set_title(f'Distribution of {col} by Churn', size=10)

    
#     handles,_=axis.get_legend_handles_labels()
#     axis.legend(handles,['lebe','gg'],loc='best')

plt.suptitle('Benchmark Distributions According to Churn Value', size=14)
plt.subplots_adjust(top=0.95,hspace=0.4,wspace=0.4)
```

###### Benchmark and Age Group

```python
fig, ax =plt.subplots(5,2, figsize=(16,18),)
for col, axis in zip(benchmark, ax.ravel()):
    per_95=np.round(np.percentile(churn[col],95))
    sns.boxplot(x='Age_group', y=col, data=churn[churn[col]<=per_95], ax=axis)
    axis.set_xlabel('')
    axis.set_title(f'Distribution of {col} by Age Group', size=10)
    
plt.suptitle('Benchmark Distributions According to Age Group', size=14)
plt.subplots_adjust(top=0.95,hspace=0.4,wspace=0.4)
```
